```
df = pd.read_csv('209TCMS.csv',encoding='unicode_escape')
```

```
# 불필요한 FORMNO, TCMSTRBLNM 컬럼 없애기
df.drop(['FORMNO','TCMSTRBLNM'], axis=1, inplace=True)
```

```
# Nan 불포함 중 겹치는 행 있는지 확인
df.duplicated().sum()
```

```
# TRBLWARNC,TRBLHEVYCD가 발생한 날짜
df['DRVDT'].unique()
```

```
# TRBLWARNC,TRBLHEVYCD가 발생한 날짜 수
pd.value_counts(df['DRVDT'].unique()).sum()
```

```
transactions = []

# 날짜별 고장코드 list로 정리
for i in range(pd.value_counts(df['DRVDT'].unique()).sum()):
    transactions.append(list(df[df['DRVDT']==df['DRVDT'].unique()[i]]['TCMSTRBLCD']))
```

```
transactions
len(transactions)
```

```
patterns = pyfpgrowth.find_frequent_patterns(transactions,34) #빈도수 34
```

```
print(patterns)
```

```
rules = pyfpgrowth.generate_association_rules(patterns, 0.9) #신뢰도 0.9
```
