```
df = pd.read_csv('209TCMS.csv',encoding='unicode_escape')
```

```
# 불필요한 FORMNO, TCMSTRBLNM 컬럼 없애기
df.drop(['FORMNO','TCMSTRBLNM'], axis=1, inplace=True)
```

```
# TRBLWARNC,TRBLHEVYCD 결측치 값 제거
TRBLWARNC_NotNull = df[df['TRBLWARNC'].notnull()]
TRBLHEVYCD_NotNull = df[df['TRBLHEVYCD'].notnull()]
TCMS_NotNull = TRBLWARNC_NotNull.append(TRBLHEVYCD_NotNull,ignore_index=True)
```

```
# Nan 불포함 중 겹치는 행 있는지 확인
TCMS_NotNull.duplicated().sum()
```

```
# TRBLWARNC,TRBLHEVYCD가 발생한 날짜
TCMS_NotNull['DRVDT'].unique()
```

```
# TRBLWARNC,TRBLHEVYCD가 발생한 날짜 수
pd.value_counts(TCMS_NotNull['DRVDT'].unique()).sum()
```

```
dataset = []

# 날짜별 고장코드 list로 정리
for i in range(pd.value_counts(TCMS_NotNull['DRVDT'].unique()).sum()):
    dataset.append(list(TCMS_NotNull[TCMS_NotNull['DRVDT']==TCMS_NotNull['DRVDT'].unique()[i]]['TCMSTRBLCD']))
```

```
dataset
len(dataset)
```

```
patterns = pyfpgrowth.find_frequent_patterns(dataset, 34) # 34이상의 빈도수
```

```
print(patterns)
```

```
rules = pyfpgrowth.generate_association_rules(patterns, 0.1) #신뢰도 0.9 이상
```
